```python

def cleaning_text(content: list):
    cleaned_text = []
    
    
    for text in content:
        lower = text.lower()
                
                # Replacing the repeating pattern of &#039;
        pattern_remove = lower.replace("&#039;", "")
                
                # Removing all the special Characters
        special_remove = pattern_remove.replace(r'[^\w\d\s]',' ')
                
                # Removing all the non ASCII characters
        ascii_remove = special_remove.replace(r'[^\x00-\x7F]+',' ')
                
                # Removing the leading and trailing Whitespaces
        whitespace_remove = ascii_remove.replace(r'^\s+|\s+?$','')
                
                # Replacing multiple Spaces with Single Space
        multiw_remove = whitespace_remove.replace(r'\s+',' ')
                
                # Replacing Two or more dots with one
        dataframe = multiw_remove.replace(r'\.{2,}', ' ')
            
        cleaned_text.append(dataframe.replace('"',''))
    
    return cleaned_text
    
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import contractions
import string
import re
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')

def filter_stop_words(text):
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))

    #This normalizes the texts by converting all the words into normal form(example, couldn't to could not)
    text= contractions.fix(text)

    #to remove the urls and normalizing the texts by removing the accent letters.
    #This also converts the text into ascii and then analyzes it
    # text = unidecode.unidecode(text)

    #Removes the punctuations
    text = text.translate(str.maketrans('', '', string.punctuation))

    #tokenizes the words
    tokens = word_tokenize(text)

    text = [word for word in tokens if word not in stop_words and len(word)>2 or re.match(r"-?\d{1,3}(,\d{3})*(\.\d+)?|\d+(\.\d+)?",word)]

    lemmatized_tokens = [lemmatizer.lemmatize(text) for text in text]
    return lemmatized_tokens
```

```python
import faiss

#indexing the documents function

def index_docs_Euclidian(doc_vec_list:list, doc_with_embed:dict):
    # as the index documents always maintain the order in which they have inserted so we can get index and then the metadata by just querying it. 
    for docs in range(len(doc_vec_list)):    
        dimension = doc_vec_list[docs].shape[1]
        index = faiss.IndexFlatL2(dimension)
        index.add(doc_vec_list[docs])
        print(f'Document{docs} of topic:{doc_with_embed[docs]['curr_topic']} is indexed')
    return index

```

```python
# using bert model
import random
import torch
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
# Apply BERT tokenization to the 'summary' column
def tokenize_summary(text):
    encoding = tokenizer.batch_encode_plus(
        [text],                 # List of input texts
        padding=True,           # Pad to the maximum sequence length
        truncation=True,        # Truncate to the maximum sequence length
        return_tensors='pt',    # Return PyTorch tensors
        add_special_tokens=True # Add CLS and SEP tokens
    )
    
    input_ids = encoding['input_ids'].tolist()[0]  # Get token IDs as a list
    attention_mask = encoding['attention_mask'].tolist()[0]  # Get attention mask as a list
    
    return input_ids, attention_mask
```

```python
tokenized_summary_list = []
attention_mask_list = []

for single_summary in cleaned_summary:
    input_ids, attention_mask = tokenize_summary(single_summary)
    tokenized_summary_list.append(input_ids)
    attention_mask_list.append(attention_mask)
```

```python
tokenized_summary_list[0]
```

```python
# Define a function to apply BERT to tokenized input
def generate_embeddings_from_tokens(input_ids, attention_mask):
    input_ids_tensor = torch.tensor([input_ids])       # Convert list to tensor
    attention_mask_tensor = torch.tensor([attention_mask])  # Convert list to tensor

    # Generate embeddings using BERT
    with torch.no_grad():
        outputs = model(input_ids_tensor, attention_mask=attention_mask_tensor)
        word_embeddings = outputs.last_hidden_state.squeeze(0)  # Shape: (sequence_length, hidden_size)

    return word_embeddings.mean(dim=0).numpy()
```

```python
embedding_of_summary = []
for iter_for_list in range(len(tokenized_summary_list)):
    embedding_of_summary.append(generate_embeddings_from_tokens(input_ids = tokenized_summary_list[iter_for_list], 
                                                                attention_mask = attention_mask_list[iter_for_list])
                               )
```
